```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Generate sample data
np.random.seed(42)

# Hypothetical scenario where means are different (ANOVA True)
group1_true = np.random.normal(loc=20, scale=5, size=100)
group2_true = np.random.normal(loc=30, scale=5, size=100)

# Hypothetical scenario where means are not different (ANOVA False)
group1_false = np.random.normal(loc=25, scale=5, size=100)
group2_false = np.random.normal(loc=25, scale=5, size=100)

# Combine data into a DataFrame
df_true = pd.DataFrame({'value': np.concatenate([group1_true, group2_true]),
                        'group': ['Group 1']*100 + ['Group 2']*100,
                        'scenario': ['ANOVA True']*200})

df_false = pd.DataFrame({'value': np.concatenate([group1_false, group2_false]),
                         'group': ['Group 1']*100 + ['Group 2']*100,
                         'scenario': ['ANOVA False']*200})

df = pd.concat([df_true, df_false])

# Perform ANOVA
anova_true = f_oneway(group1_true, group2_true)
anova_false = f_oneway(group1_false, group2_false)

print(f"ANOVA True: F={anova_true.statistic}, p={anova_true.pvalue}")
print(f"ANOVA False: F={anova_false.statistic}, p={anova_false.pvalue}")

# Plot distributions
plt.figure(figsize=(14, 7))

# Plot for ANOVA True
plt.subplot(1, 2, 1)
sns.histplot(df[df['scenario'] == 'ANOVA True'], x='value', hue='group', kde=True)
plt.title('ANOVA True: Means are Different')
plt.axvline(np.mean(group1_true), color='blue', linestyle='--')
plt.axvline(np.mean(group2_true), color='orange', linestyle='--')

# Plot for ANOVA False
plt.subplot(1, 2, 2)
sns.histplot(df[df['scenario'] == 'ANOVA False'], x='value', hue='group', kde=True)
plt.title('ANOVA False: Means are Not Different')
plt.axvline(np.mean(group1_false), color='blue', linestyle='--')
plt.axvline(np.mean(group2_false), color='orange', linestyle='--')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

# Generate sample data with no intersection
np.random.seed(42)

# Group 1: Mean 20, Standard Deviation 2
group1_extreme = np.random.normal(loc=20, scale=2, size=100)
# Group 2: Mean 40, Standard Deviation 2
group2_extreme = np.random.normal(loc=40, scale=2, size=100)

# Combine data into a DataFrame
df_extreme = pd.DataFrame({
    'value': np.concatenate([group1_extreme, group2_extreme]),
    'group': ['Group 1']*100 + ['Group 2']*100
})

# Perform ANOVA
anova_extreme = f_oneway(group1_extreme, group2_extreme)
print(f"ANOVA Extreme: F={anova_extreme.statistic}, p={anova_extreme.pvalue}")

# Plot distributions
plt.figure(figsize=(14, 7))

# Histogram and KDE plot for extreme condition
sns.histplot(df_extreme, x='value', hue='group', kde=True)
plt.title('Extreme Condition: No Intersection Between Groups')
plt.axvline(np.mean(group1_extreme), color='blue', linestyle='--', label='Mean Group 1')
plt.axvline(np.mean(group2_extreme), color='orange', linestyle='--', label='Mean Group 2')
plt.legend()

plt.tight_layout()
plt.show()
```
