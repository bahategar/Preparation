For this exercise, we will be using one of the publicly available datasets from the UCI Machine Learning Repository, which can be found at https://archive.ics.uci.edu/ml/datasets/bank+marketing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import tree
```

```python
df = pd.read_csv('./data/bank/bank-full.csv', sep=';')
print(df.info())
df.head()
```

```python
# Conversion rate
df['conversion'] = df['y'].apply(lambda x: 0 if x =='no' else 1)

conversion_rate = pd.DataFrame(df.groupby('conversion').count()['y'] / df.shape[0] * 100)
conversion_rate.T
```

```python
# Conversion rate by Job

groupby_job = df.groupby('job')['conversion']
conversion_rate_by_job = groupby_job.sum() / groupby_job.count() * 100
conversion_rate_by_job
```

```python
ax = conversion_rate_by_job.plot(kind='barh', color='skyblue', 
                                 grid=True, title='Conversion Rate by Job')

ax.set_xlabel('Conversion rate (%)')
ax.set_ylabel('Job')
plt.show()
```

NOTE:
- The student and retired groups are the two groups with the highest conversion rates.
- The blue-collar and entrepreneur groups are the two groups with the lowest conversion rates.


```python
# Default rates by conversions

default_by_conversion = pd.pivot_table(df, 
                                       values='y', 
                                       index='default', 
                                       columns='conversion',
                                      aggfunc=len).fillna(0)
default_by_conversion
```

```python
default_by_conversion.plot(kind='pie', 
                           figsize=(15, 7), 
                           startangle=90, 
                           subplots=True,
                          autopct=lambda x: '%0.1f%%' % x)
plt.show()
```

NOTE: 
Although the overall percentage of the previous default is low in both groups, the default rate in the non-conversion group is about twice as high as the conversion group.

```python
# Bank balances by conversion

ax = df[['conversion', 'balance']].boxplot(by='conversion', 
                                           showfliers=False, 
                                           figsize=(10, 7))

ax.set_xlabel('Conversion')
ax.set_ylabel('Average Bank Balance')
ax.set_title('Average Bank Balance Distributions by Conversion')
plt.suptitle('')
plt.show()
```

The median of the bank balance is slightly higher for the conversion group, as compared to the non-conversion group. Also, the bank balances of converted customers seem to vary more than those of non-converted customers.

```python
# Conversion rates by number of contacts
groupby_contacts = df.groupby('campaign')['conversion']

conversions_by_num_contacts = groupby_contacts.sum() / groupby_contacts.count() * 100
pd.DataFrame(conversions_by_num_contacts).head()
```

```python
ax = conversions_by_num_contacts.plot(kind='bar', 
                                      figsize=(10, 7), 
                                      title='Conversion Rate by Number of Contacts',
                                      grid=True,    
                                      color='skyblue'
                                     )

ax.set_xlabel('Number of Contacts')
ax.set_ylabel('Conversion Rate (%)')
plt.show()
```

NOTE: 
There's some noise in a higher numbers of contacts, but the overall downward trend in this bar chart. As the number of contacts increases, the conversion rates slowly decrease.

```python
# Encoding categorical variables

# Encoding month
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df['month'] = df['month'].apply(lambda x: months.index(x)+1)

# Encoding job
jobs_encoded = pd.get_dummies(df['job'])
jobs_encoded.columns = ['job_%s' % x for x in jobs_encoded.columns]
df = pd.concat([df, jobs_encoded], axis=1)

# Encoding marital
marital_encoded = pd.get_dummies(df['marital'])
marital_encoded.columns = ['marital_%s' % x for x in marital_encoded.columns]
df = pd.concat([df, marital_encoded], axis=1)

# Encoding housing
df['housing'] = df['housing'].apply(lambda x: 1 if x == 'yes' else 0)

# Encoding loan
df['loan'] = df['loan'].apply(lambda x: 1 if x == 'yes' else 0)

```

```python
# Modeling

features = [
    'age',
    'balance',
    'campaign',
    'previous',
    'housing',
] + list(jobs_encoded.columns) + list(marital_encoded.columns)

response_var = 'conversion'

features
```

```python
dt_model = tree.DecisionTreeClassifier(max_depth=4)
dt_model.fit(df[features], df[response_var])
```

```python
dt_model.classes_
```

```python
# Interpreting Decision Tree Model

from sklearn.tree import plot_tree

plt.figure(figsize=(30, 15))
plot_tree(dt_model, filled=True, feature_names=features, class_names=response_var)
plt.show()
```
