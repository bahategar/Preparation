```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Using Tree-Based Regression Models to Capture Non-Linear Trends

The goal:
> Figuring out how spending habits change with age in your customers, and how well different models can capture this relationship.

NOTE:

Having a model like this can help in building age-specific website patterns for your customers since you will be able to recommend products that suit the customers' spend levels.

```python
df = pd.read_csv('./dataset/age_spend.csv')

print(df.info())
df.head()
```

```python
from sklearn.model_selection import train_test_split

X = df[['age']]
y = df['spend']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
# Non-linear model
from sklearn.tree import DecisionTreeRegressor


max2_tree_model = DecisionTreeRegressor(max_depth=2)
max2_tree_model.fit(X_train, y_train)

max5_tree_model = DecisionTreeRegressor(max_depth=5)
max5_tree_model.fit(X_train, y_train)
```

```python
# Linear model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

linear_predictions = model.predict(X_test)
rmse_linear_model = mean_squared_error(linear_predictions, y_test)**0.5
print(f"Lineaer model RMSE: {rmse_linear_model}")

max2_tree_predictions = max2_tree_model.predict(X_test)
rmse_max2_tree_model = mean_squared_error(max2_tree_predictions, y_test)**0.5
print(f"Lineaer model RMSE: {rmse_max2_tree_model}")

max5_tree_predictions = max5_tree_model.predict(X_test)
rmse_max5_tree_model = mean_squared_error(max5_tree_predictions, y_test)**0.5
print(f"Lineaer model RMSE: {rmse_max5_tree_model}")
```

NOTE:
Notice that the linear model has the largest error, the decision tree with a
maximum depth of 2 does better, and the decision tree with a maximum depth
of 5 has the lowest error of the three.

```python
ages = pd.DataFrame({'age':range(18,70)})

plt.scatter(X_test.age.tolist(), y_test.tolist(), color='gray')
plt.plot(ages, model.predict(ages), 
         color='k', linewidth=5, label='Linear Regression')
plt.xlabel('Age')
plt.ylabel('Spend')
plt.show()
```

```python
plt.scatter(X_test.age.tolist(), y_test.tolist(), color='gray')
plt.plot(ages, max2_tree_model.predict(ages), 
         color='k', linewidth=5, label='Tree with max depth 2')
plt.xlabel('Age')
plt.ylabel('Spend')
plt.show()
```

```python
plt.scatter(X_test.age.tolist(), y_test.tolist(), color='gray')
plt.plot(ages, max5_tree_model.predict(ages), 
         color='k', linewidth=5, label='Tree with max depth 5')
plt.xlabel('Age')
plt.ylabel('Spend')
plt.show()
```

NOTES:
- You can see that the linear regression model just shows a flat line across the ages; it is unable to capture the fact that people aged around 40 spend more, while people younger and older than 40 spend less.
- The max 2 depth model does a better job of capturing the relationship, though it does not capture the sharp decline in the oldest or youngest populations.
- The max 5 depth tree model does an even better job of capturing the relationship, properly capturing a sharp decline in the oldest or the youngest population.

```python
# Try using Random Forest Model
from sklearn.ensemble import RandomForestRegressor

max2_forest_model = RandomForestRegressor(max_depth=2, random_state=10)
max2_forest_model.fit(X_train, y_train)

max5_forest_model = RandomForestRegressor(max_depth=5, random_state=10)
max5_forest_model.fit(X_train, y_train)
```

```python
max2_forest_predictions = max2_forest_model.predict(X_test)
rmse_max2_forest_model = mean_squared_error(max2_forest_predictions, y_test)**0.5
print(f"Lineaer model RMSE: {rmse_max2_forest_model}")

max5_forest_predictions = max5_forest_model.predict(X_test)
rmse_max5_tree_model = mean_squared_error(max5_forest_predictions, y_test)**0.5
print(f"Lineaer model RMSE: {rmse_max5_tree_model}")
```

```python
plt.scatter(X_test.age.tolist(), y_test.tolist(),color='gray')
plt.plot(ages,max2_forest_model.predict(ages), color='k',\
linewidth=5, label="Forest with max depth 2")
plt.xlabel("age")
plt.ylabel("spend")
plt.show()
```

```python
plt.scatter(X_test.age.tolist(), y_test.tolist(), color='gray')
plt.plot(ages,max5_forest_model.predict(ages), color='k',\
linewidth=5, label="Forest with max depth 5")
plt.xlabel("age")
plt.ylabel("spend")
plt.show()
```

NOTES:
- We can see that the max 2 depth forest model captures the data trend better than the decision tree, but still doesn't quite capture the trend at the very high or low ends of our range.
- Again, in the forest model, the greater maximum depth does an even better job of capturing the relationship, properly capturing the sharp decline in the oldest and youngest population groups.

```python
plt.figure(figsize=(12,8))
plt.scatter(X_test.age.tolist(), y_test.tolist())
plt.plot(ages,model.predict(ages), color='r', linewidth=5, \
label="Linear Regression")
plt.plot(ages,max2_tree_model.predict(ages), color='g',\
linewidth=5,label="Tree with max depth 2")
plt.plot(ages,max5_tree_model.predict(ages), color='k',\
linewidth=5, label="Tree with max depth 5")
plt.plot(ages,max2_forest_model.predict(ages), color='c',\
linewidth=5, label="Forest with max depth 2")
plt.plot(ages,max5_forest_model.predict(ages), color='m',\
linewidth=5, label="Forest with max depth 5")
plt.legend()
plt.xlabel("age")
plt.ylabel("spend")
plt.show()
```
