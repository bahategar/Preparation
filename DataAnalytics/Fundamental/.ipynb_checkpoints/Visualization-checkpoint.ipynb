```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('classic')
```

# Introduction

> In Matplotlib, the figure (an instance of the class plt.Figure) can be tought of as a single container that contains all the objects representing axes, graphics, text, and labels.

> The axes (an instance of the class plt.Axes) is a bounding box with ticks and labels, which will eventually contain the plot elements that make up our visualization.

# Simple Line Plots

```python
fig = plt.figure()
ax = plt.axes()

print(50 * '=')
print("GENERATE SIMPLE PLOT")
print(50 * '=')

# Alternative 1
x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x))
ax.plot(x, np.cos(x))
plt.show()

# Alternative 2: Directly plot with plt.plot()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()

print(50 * '=')
print("Adjusting the Plot: Line Colors and Styles")
print(50 * '=')


x = np.linspace(0, 10, 100)
plt.plot(x, x + 0, color='blue', linestyle='solid') # eq: linestyle='-'
plt.plot(x, x + 1, color='cyan', linestyle='dashed') # eq: linestyle='--'
plt.plot(x, x + 2, color='black', linestyle='dashdot') # eq: linestyle='-.'
plt.plot(x, x + 3, color='red', linestyle='dotted') # eq: linestyle=':'
plt.show()
```

# Simple Scatter Plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

# Alternative 1
plt.plot(x, y, 'o', color='black') # 'o-' format = scatter-line; 'o' format = scatter
plt.show()

# Alternative 2
plt.scatter(x, y, color='red', )
plt.show()
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, )
plt.colorbar() # Show color scale
plt.show()
```
