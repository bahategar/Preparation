```python
# Import package for getting dataset example
import wooldridge as woo

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf

import math
from tqdm import tqdm
```

# Introduction

**Concept:**
> In case model that follows Gauss-Markov Assumptions, The distribution of residu is asymptotically normal as the size sample increases. The consequent then the estimator also asymptotically normal.

NOTE:
- In small samples, the estimator have a normal sampling distribution only if: (1) The error terms are normally distributed and (2) We condition on the regressors.
- The regressors does not have to normally distribution. The assumption of model do not tell about distribution of regressors.

### Normally Distributed Error Terms
**Conditions:**
- The error terms as normally distributed
- We condition on the regressors (by make it as fixed)

```python
def sampling_estimator_1(n, beta0=1, beta1=0.5,
                         std_x=1, mean_x=4, r=1000,
                         random_state=1234567):
    # Set the random seed
    np.random.seed(random_state)
    
    # Initialize b1 to store results
    b = np.empty(r)
    
    # Draw a sample of x, fixed over replications
    x = stats.norm.rvs(mean_x, std_x, size=n)
    
    # Repeat r times
    for i in tqdm(range(r), desc="Iterations"):
        # Draw a sample of u (std. normal)
        u = stats.norm.rvs(0, 1, size=n)
        y = beta0 + beta1 * x + u
        df = pd.DataFrame({'y': y, 'x': x})

        # Estimate conditional OLS
        model = smf.ols(formula='y ~ x', data=df).fit()
        b[i] = model.params['x']
    
    return b
    
def plot_distribution(b, n, ax):
    # mean and std
    mean_data = np.mean(b)
    std_data = np.std(b, ddof=1)

    # Estimate kernel density
    kde = sm.nonparametric.KDEUnivariate(b)
    kde.fit()

    # Plot the kernel density result on the given axis
    ax.plot(kde.support, kde.density, color="black", lw=2, label='b')

    # Plot the theoretical normal distribution
    x = np.linspace(min(b), max(b), 1000)
    ax.plot(x, stats.norm.pdf(x, mean_data, std_data), 'k--', lw=2, label='normal distribution')

    # Set title and labels
    ax.set_title(f"Sample Size: {n}")
    ax.set_ylabel('Density')
    ax.legend()

# Example usage:
samples = [5, 10, 30, 50, 75, 100]
fig, axes = plt.subplots(len(samples), 1, figsize=(10, 18))

for ax, n in zip(axes, samples):
    b = sampling_estimator_1(n, r=1000)
    plot_distribution(b, n, ax)

plt.tight_layout()
plt.show()
```

### Non-Normal Error Terms

**Conditions:**
- The error term is not distributed normal
- We condition on the regressors (by make it as fixed)

NOTE: If the distribution of the error terms is not as exteremely non-normal, smaller sample sizes like the rule of thumb n = 30 might suffice for valid asymptotics.

```python
def sampling_estimator_2(n, beta0=1, beta1=0.5,
                         std_x=1, mean_x=4,
                         r=1000, random_state=1234567):
    # Set the random seed
    np.random.seed(random_state)
    
    # Initialize b1 to store results
    b = np.empty(r)
    
    # Draw a sample of x, fixed over replications
    x = stats.norm.rvs(mean_x, std_x, size=n)
    
    # Repeat r times
    for i in tqdm(range(r), desc="Iterations"):
        # Draw a sample of u (standardized chi-squared[1])
        u = (stats.chi2.rvs(1, size=n) - 1) / np.sqrt(2)
        y = beta0 + beta1 * x + u
        df = pd.DataFrame({'y': y, 'x': x})

        # Estimate conditional OLS
        model = smf.ols(formula='y ~ x', data=df).fit()
        b[i] = model.params['x']
    
    return b
    
def plot_distribution(b, n, ax):
    # mean and std
    mean_data = np.mean(b)
    std_data = np.std(b, ddof=1)

    # Estimate kernel density
    kde = sm.nonparametric.KDEUnivariate(b)
    kde.fit()

    # Plot the kernel density result on the given axis
    ax.plot(kde.support, kde.density, color="black", lw=2, label='b')

    # Plot the theoretical normal distribution
    x = np.linspace(min(b), max(b), 1000)
    ax.plot(x, stats.norm.pdf(x, mean_data, std_data), 'k--', lw=2, label='normal distribution')

    # Set title and labels
    ax.set_title(f"Sample Size: {n}")
    ax.set_ylabel('Density')
    ax.legend()

# Example usage:
samples = [5, 10, 30, 50, 75, 100]
fig, axes = plt.subplots(len(samples), 1, figsize=(10, 18))

for ax, n in zip(axes, samples):
    b = sampling_estimator_2(n, r=1000)
    plot_distribution(b, n, ax)

plt.tight_layout()
plt.show()
```

### (Not) Conditioning on the Regressors

**Conditions:**
- The error term is distributed normal
- We do not condition on the regressors (by make it not fixed)

```python
def sampling_estimator_3(n, beta0=1, beta1=0.5,
                         std_x=1, mean_x=4, r=1000,
                         random_state=1234567):
    # Set the random seed
    np.random.seed(random_state)
    
    # Initialize b1 to store results
    b = np.empty(r)
        
    # Repeat r times
    for i in tqdm(range(r), desc="Iterations"):
        # Draw a sample of x
        x = stats.norm.rvs(mean_x, std_x, size=n)

        # Draw a sample of u (std. normal)
        u = stats.norm.rvs(0, 1, size=n)
        y = beta0 + beta1 * x + u
        df = pd.DataFrame({'y': y, 'x': x})

        # Estimate conditional OLS
        model = smf.ols(formula='y ~ x', data=df).fit()
        b[i] = model.params['x']
    
    return b
    
def plot_distribution(b, n, ax):
    # mean and std
    mean_data = np.mean(b)
    std_data = np.std(b, ddof=1)

    # Estimate kernel density
    kde = sm.nonparametric.KDEUnivariate(b)
    kde.fit()

    # Plot the kernel density result on the given axis
    ax.plot(kde.support, kde.density, color="black", lw=2, label='b')

    # Plot the theoretical normal distribution
    x = np.linspace(min(b), max(b), 1000)
    ax.plot(x, stats.norm.pdf(x, mean_data, std_data), 'k--', lw=2, label='normal distribution')

    # Set title and labels
    ax.set_title(f"Sample Size: {n}")
    ax.set_ylabel('Density')
    ax.legend()

# Example usage:
samples = [5, 10, 30, 50, 75, 100]
fig, axes = plt.subplots(len(samples), 1, figsize=(10, 18))

for ax, n in zip(axes, samples):
    b = sampling_estimator_3(n, r=1000)
    plot_distribution(b, n, ax)

plt.tight_layout()
plt.show()
```

# Lagrange Multiplier Testing


**Define Problems**

Let we define:
- Unrestricted model with $k$ independent: $y_{ur} = \beta^{ur}_{0} + \beta^{ur}_{1}x_1 + \dots + \beta^{ur}_{k}x_k + e_{ur}$
- Restricted model with $q$ exclusion restrictions: $y_{r} = \beta^{r}_{0} + \beta^{r}_{1}x_1 + \dots + \beta^{r}_{k-q}x_{k-q} + e_{r}$ 

Then we have LM:
$$
LM = n R^2_{u_r}
$$
Where,
- $R^2_{u_r} = $ R-Squared regression $u_r$ on all regressors on unrestricted model
- $n$ = Number of samples

NOTE: 
- The distribution of the LM test statistic under the null hypothesis approaches a chi-square distribution $\chi^2_{q}$ (where $q$ is number of exclusion regressors).
- This is alternative F-testing.
- Also it works only when size sample large. If sample is not large, the distribution test statistic might not follow chi-square distribution.


**Hypothesis Testing**

![image](images/5_LM-table1.png)

![image](images/Example_5-3.png)

```python
crime1 = woo.dataWoo('crime1')

# 1. Estimate Restricted Model
model_r = smf.ols(formula='narr86 ~ pcnv + ptime86 + qemp86', data=crime1).fit()
r2_r = model_r.rsquared
print(f"R-squared restricted = {r2_r}")

# 2. Regression of Residuals from Restricted Model
crime1['error_r'] = model_r.resid
model_er = smf.ols(formula='error_r ~ pcnv + ptime86 + qemp86 + avgsen + tottime',
                   data=crime1).fit()
r2_LM = model_er.rsquared
print(f"R-squared Residuals Restricted on Regressors = {r2_LM}")

# 3. Calculation of LM test statistic
LM = r2_LM * model_er.nobs
print(f"LM = {LM}")

# 4. Critical Value from Chi-squared distribution, alpha=10%
alpha = .1
q = 2
chi_square_critical = stats.chi2.ppf(1 - alpha, q)
print(f"Critical value = {chi_square_critical}")

# 5. p value (alternative to critical value)
pval = 1 - stats.chi2.cdf(LM, q)
print(f"p-value = {pval}")

# 6 Compare to F-test
model = smf.ols(formula='narr86 ~ pcnv + ptime86 + qemp86 + avgsen + tottime',
                data=crime1).fit()
hypotheses = ['avgsen = 0', 'tottime = 0']
ftest = model.f_test(hypotheses)
fstat = ftest.statistic
fpval = ftest.pvalue
print(f"fstat = {fstat}")
print(f"fpval = {fpval}")
```
