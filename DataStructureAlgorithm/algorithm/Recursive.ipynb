```python
# String Reversal

def reverseString(text):
    if (text == ""):
        return ""
    return reverseString(text[1:]) + text[0]

print(reverseString("Hello"))
```

```python
# Palindrome
def isPalindrome(text):
    if (len(text) <= 1):
        return True
    if(text[0] == text[-1]):
        return isPalindrome(text[1:-1])    
    return False

print(isPalindrome('racecar'))
```

```python
# Sum of Natural Numbers
def recursiveSum(n):
    if (n <= 1):
        return n
    return n + recursiveSum(n - 1)

print(recursiveSum(4))
```

```python
# Binary search
def binarySearch(array, left, right, x):
    if (left > right):
        return -1
    
    mid = (left + right) // 2
    
    if (x == array[mid]):
        return mid
    
    if (x < array[mid]):
        return binarySearch(array, left, mid - 1, x)
    
    return binarySearch(array, mid + 1, right, x)


arr = [2, 3, 4, 10, 40]
x = 2
result = binarySearch(arr, 0, len(arr) - 1, x)
print(result)
```

```python
# Fibonacci
def fib(n):
    if ((n == 0) or (n == 1)):
        return n
    else:
        return fib(n - 1) + fib(n - 2)
    
fib(4)
```

```python
# Problem:
# Explain the functionality of the following functions.

def fun1(x, y) : 
  
    if (x == 0) : 
        return y 
    else : 
        return fun1(x - 1, x + y) 
    
# Tracing:
# x = 0
# fun(0, y) -> y

# x = 1
# fun(1, y) -> fun(0, 1 + y) -> 1 + y

# x = 2
# fun(2, y) -> fun(1, 2 + y) -> fun(0, 1 + 2 + y) -> 1 + 2 + y

# x = 3
# fun(3, y) -> fun(2, 3 + y) -> fun(1, 2 + 3 + y) -> fun(0, 1 + 2 + 3 + y) -> 1 + 2 + 3 + y

# It will return [(x + 1) * x]/2 + y
# The time complexity of this function: O(x)
# The space complexity of this function: O(x) -> Stack frames

# If x = 5 and y = 2 -> 17

print(fun1(5, 2))
```

```python
# Question
# Predict the output of the following program. 
# What does the following fun() do in general?  

   
def fun(a, b):
    if (b == 0):
        return 0
    if (b % 2 == 0):
        return fun(a + a, b//2) 
     
    return fun(a + a, b//2) + a 
 
# Tracer
# b = 0
# f(a, 0) => 0

# b = 1
# f(a, 1) => fun(2a, 0) + a => 0 + a

# b = 2
# f(a, 2) => fun(2a, 1) => fun(4a, 0) + 2a => 0 + 2a

# b = 3
# f(a, 3) => fun(2a, 2) + a => [fun(4a, 0) + 2a] + a => 0 + 3a

# The function will return b*a
# The time complexity: O(log a*b)
# The space complexity: O(log a*b)

# Prediction of fun(4, 3) => 12.

# Driver code
print(fun(4, 3))
```

```python
# Question
# Predict the output of the following program. What does the following 
#  fun() do in general? 

fp = 15
def fun(n): 
    global fp 
    if (n <= 2): 
        fp = 1
        return 1
  
    t = fun(n - 1) 
    f = t + fp 
    fp = t 
    return f 

# Tracer
# n = 2
# fun(2) => 1; fp=1

# n = 3
# fun(3) => fun(2) => 1; fp = 1 => 2; fp = 1

# n = 4
# fun(4) => fun(3)[rec] => fun(2) => 2; fp = 1 => 2 + 1; fp = 2;

# n = 5
# fun(5) => fun(4)[rec] => fun(3) => fun(2) => 
#    2; fp = 1 => 3; fp = 2 => 3 + 2; fp=3

# n = 6
# fun(6) => fun(5)[rec] => fun(4) => fun(3) => fun(2) => 
#    2; fp = 1 => 3; fp = 2 => 5; fp=3 => 5 + 3; fp = 5

# 2 -> 1 (1)
# 3 -> 2 (1)
# 4 -> 3 (2)
# 5 -> 5 (3)
# 6 -> 8 (5)

# Conclusion:
# The function will return the n+1 number of Fibonacci series.
#  Fp follows Fibonacci series.

# Fibonacci series:
# 0, 1, 1, 2, 3, 5, 8, 13

# Driver code 
print(fun(7))
```

```python
# Summing Numbers in an Array

def sum(numbers):
    if (len(numbers) == 0): 
        # BASE CASE
        return 0
    else:
        return numbers[0] + sum(numbers[1:])

test1 = [1, 2, 3, 4, 5]
print(f"The sum of {test1} is ", sum(test1))

test1 = [5, 2, 4, 8]
print(f"The sum of {test1} is ", sum(test1))

test1 = [1, 10, 100, 1000]
print(f"The sum of {test1} is ", sum(test1))
```

```python
# Reversing a String

# CAT => TAC

def reverse(theString):
    if len(theString) == 0:
        return ""
    return reverse(theString[1:]) + theString[0]

print(reverse("CAT"))
```

```python
# Print start pyramid

def reverseStar(k, n):
    if (k == n):
        print("*" * k, end="\n")
        return
    print('*' * k, end="\n")
    reverseStar(k + 1, n)
    print('*' * k, end="\n")
    

reverseStar(0, 10)
```

```python
# is palindrome

def isPalindrome(text):
    trimText = text.strip()
    if (len(trimText) == 0) or (len(trimText) == 1):
        return True
    
    if (trimText[0].lower() == trimText[-1].lower()):
        start = 1
        last = len(trimText) - 1
        return isPalindrome(trimText[start:last])
    return False
```

```python
# Tower of Hanoi

def towerOfHanoi(k, from_rod, to_rod, aux_rod):
    if k == 0:
        return
    
    towerOfHanoi(k - 1, from_rod, aux_rod, to_rod)
    print(f"Move disk {k} from {from_rod} to {to_rod}")
    towerOfHanoi(k - 1, aux_rod, to_rod, from_rod)
    
towerOfHanoi(3, "A", "C", "B")
```
