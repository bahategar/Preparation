```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

import pandas as pd
```

```python
url = 'https://www.audible.com/search'
path = '../driver/chromedriver.exe'

service = Service(path)
```

```python
# Without pagination

import time
# main_box xpath: //div[@id="center-3"]//ul[contains(@class, "bc-list")]/li[contains(@class, "title")]

title = list()
author = list()
narrator = list()
runtime = list()
language = list()
ratings = list()
mostPopular = list()
trending = list()
priceLine = list()

map_frame = {
    'title': title,
    'author': author,
    'narrator': narrator,
    'runtime': runtime,
    'language': language,
    'ratings': ratings,
    'mostPopular': mostPopular,
    'trending': trending,
    'priceLine': priceLine
}

def storage_function(storage, text):
    if text == 'title':
        result = box.find_element(by='xpath',
                                  value=f'.//h3').text
        storage.append(result)
    else:
        try:
            result = box.find_element(by='xpath',
                                      value=f'./li[contains(@class, "{text}")]').text
            storage.append(result)
        except NoSuchElementException:
            storage.append("")

# Create bot
driver = webdriver.Chrome(service=service)

driver.get(url)

main_box = WebDriverWait(driver, 10).until(EC.visibility_of_any_elements_located((By.XPATH, '//div[@id="center-3"]//ul[contains(@class, "bc-list")]')))

if len(main_box) == 0:
    time.sleep(3)
    main_box = driver.find_elements(by='xpath', value='//div[@id="center-3"]//ul[contains(@class, "bc-list")]')

        
for box in main_box:
    storage_function(title, 'title')
    storage_function(author, 'author')
    storage_function(narrator, 'narrator')
    storage_function(runtime, 'runtime')
    storage_function(language, 'language')
    storage_function(ratings, 'ratings')
    storage_function(mostPopular, 'mostPopular')
    storage_function(trending, 'trending')
    storage_function(priceLine, 'priceLine')
    
driver.quit()
df = pd.DataFrame(map_frame)
df.head()
```

```python
# With pagination

import time
# main_box xpath: //div[@id="center-3"]//ul[contains(@class, "bc-list")]/li[contains(@class, "title")]

title = list()
author = list()
narrator = list()
runtime = list()
language = list()
ratings = list()
mostPopular = list()
trending = list()
priceLine = list()

map_frame = {
    'title': title,
    'author': author,
    'narrator': narrator,
    'runtime': runtime,
    'language': language,
    'ratings': ratings,
    'mostPopular': mostPopular,
    'trending': trending,
    'priceLine': priceLine
}

def storage_function(storage, text):
    if text == 'title':
        result = box.find_element(by='xpath',
                                  value=f'.//h3').text
        storage.append(result)
    else:
        try:
            result = box.find_element(by='xpath',
                                      value=f'./li[contains(@class, "{text}")]').text
            storage.append(result)
        except NoSuchElementException:
            storage.append("")

# Create bot
driver = webdriver.Chrome(service=service)

driver.get(url)

main_box = WebDriverWait(driver, 10)\
                .until(EC.visibility_of_any_elements_located(
                        (By.XPATH, '//div[@id="center-3"]//ul[contains(@class, "bc-list")]')))
    
if len(main_box) == 0:
    time.sleep(3)
    main_box = driver.find_elements(by='xpath', value='//div[@id="center-3"]//ul[contains(@class, "bc-list")]')

# Pagination
pagination = driver.find_element(by='xpath', value='//ul[contains(@class, "pagingElements")]')
last_page = int(pagination.find_elements(by='xpath', value='./li')[-2].text)
current_page = 1

while current_page <= last_page // 4:
    if current_page > 1:
        driver.get(f'{url}?page={current_page}')
        main_box = WebDriverWait(driver, 10)\
                    .until(EC.visibility_of_any_elements_located(
                                (By.XPATH, '//div[@id="center-3"]//ul[contains(@class, "bc-list")]')))
    
        if len(main_box) == 0:
            time.sleep(3)
            main_box = driver.find_elements(by='xpath', value='//div[@id="center-3"]//ul[contains(@class, "bc-list")]')
        
    for box in main_box:
        storage_function(title, 'title')
        storage_function(author, 'author')
        storage_function(narrator, 'narrator')
        storage_function(runtime, 'runtime')
        storage_function(language, 'language')
        storage_function(ratings, 'ratings')
        storage_function(mostPopular, 'mostPopular')
        storage_function(trending, 'trending')
        storage_function(priceLine, 'priceLine')
    
    current_page += 1
    
driver.quit()
df = pd.DataFrame(map_frame)
df.head()
```
