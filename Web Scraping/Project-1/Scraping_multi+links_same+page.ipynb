```python
import requests
from bs4 import BeautifulSoup
```

```python
root = 'https://subslikescript.com'
target = f'{root}/movies'

result = requests.get(target)
content = result.text
soup = BeautifulSoup(content, 'lxml')
print(soup.prettify())
```

```python
# Get all links
hyperlink_tags = soup.find(name='ul', attrs={'class':'scripts-list'}).find_all(name='a')
links = []
for hyperlink in hyperlink_tags:
    links.append(root+hyperlink['href'])
```

```python
for link in links:
    request_script_page = requests.get(f'{link}')
    soup_script = BeautifulSoup(request_script_page.text, 'lxml')
    text = soup_script.find(name='div', attrs={'class': 'full-script'}).get_text(strip=True, separator=' ')
    title = soup_script.find(name='article', attrs={'class': 'main-article'}).h1.text.replace(' ', '_')
    
    with open(f'./result/{title}.txt', 'w', encoding='utf-8') as file:
        file.write(title)
        file.write('\n')
        file.write(text)
```
